```python
import numpy as np

# 수치 미분 함수
def numerical_derivative(f, x):
    delta_x = 1e-4
    grad = np.zeros_like(x)
    
    it = np.nditer(x, flags=['multi_index'], op_flags=['readwrite'])
    
    while not it.finished:
        idx = it.multi_index
        tmp_val = x[idx]
        x[idx] = float(tmp_val) + delta_x
        fx1 = f(x)
        
        x[idx] = tmp_val - delta_x
        fx2 = f(x)
        grad[idx] = (fx1 - fx2) / (2 * delta_x)
        
        x[idx] = tmp_val
        it.iternext()
        
    return grad
        
# calculation function
def predict(x):
    y = np.dot(x, W) + b
    return y

# loss function : mean square error
def loss_func(x, t):
    y = np.dot(x, W) + b
    return ( np.sum((t-y)**2) ) / (len(x))

# data read
loaded_data = np.loadtxt('./regression_testdata_03.csv', delimiter=',', dtype=np.float32)

x_data = loaded_data[ : , 0 : -1]
t_data = loaded_data[ : , [-1]]

print('x shape : ', x_data.shape)
print('t shape : ', t_data.shape)

W = np.random.rand(4,1)
b = np.random.rand(1)

print('W shape : ', W.shape)
print('b shape : ', b.shape)

learning_rate = 1e-2

f = lambda x : loss_func(x_data, t_data)

print('inital error value = ', loss_func(x_data, t_data), ', inital W = ', W, 'b = ', b)

for step in range(8001):
    W -= learning_rate * numerical_derivative(f, W)
    b -= learning_rate * numerical_derivative(f, b)
    
    if( step % 400 == 0):
        print('step = ', step, ', error value = ', loss_func(x_data, t_data), ', W = ', W, ', b = ', b)

predict(np.array([1,2,3,4]))
```
